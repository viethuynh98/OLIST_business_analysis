```python
import pandas as pd

# Đọc dữ liệu từ file CSV
order_items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')

# Kết hợp dữ liệu
merged_data = pd.merge(order_items, products, on='product_id')

# print(merged_data.head())

# Phân tích mối quan hệ giữa weight_g và freight_value
import matplotlib.pyplot as plt

plt.scatter(merged_data['product_weight_g'], merged_data['freight_value'])
plt.xlabel('Product Weight (g)')
plt.ylabel('Freight Value')
plt.title('Relationship between Product Weight and Freight Value')
plt.show()

# Phân tích mối quan hệ giữa kích thước sản phẩm (chiều dài, chiều rộng, chiều cao) và freight_value
merged_data['product_volume_cm3'] = merged_data['product_length_cm'] * merged_data['product_height_cm'] * merged_data['product_width_cm']

plt.scatter(merged_data['product_volume_cm3'], merged_data['freight_value'])
plt.xlabel('Product Volume (cm³)')
plt.ylabel('Freight Value')
plt.title('Relationship between Product Volume and Freight Value')
plt.show()
```
