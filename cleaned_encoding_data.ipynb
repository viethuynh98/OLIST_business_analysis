```python
import chardet

file_path = '/content/drive/MyDrive/1_Python_practices/final_project_data/olist_order_reviews_dataset.csv'

# Phát hiện mã hóa của file
with open(file_path, 'rb') as f:
    result = chardet.detect(f.read(100000))  # Đọc 10000 byte đầu tiên
    encoding = result['encoding']
    print(f"Encoding detected: {encoding}")
```

```python
import pandas as pd
import unicodedata

# Đọc file CSV với mã hóa xác định
df = pd.read_csv(file_path, encoding=encoding)

# Hàm làm sạch ký tự đặc biệt và chuẩn hóa Unicode
def clean_and_normalize_text(text):
    if pd.isna(text):
        return text
    # Chuẩn hóa Unicode
    normalized_text = unicodedata.normalize('NFKD', text)
    # Loại bỏ các ký tự có thể được sử dụng để tạo icon
    cleaned_text = ''.join(c for c in normalized_text if unicodedata.category(c) not in ['So', 'Sk', 'Sm', 'Sc', 'Pc', 'Pi', 'Pf', 'Ps', 'Pe'])
    # Thay thế các ký tự xuống dòng bằng dấu cách
    cleaned_text = cleaned_text.replace('\n', ' ').replace('\r', ' ')
    return cleaned_text.strip()

# Áp dụng hàm làm sạch và chuẩn hóa cho cột 'review_comment_message_en'
df['review_comment_message_en'] = df['review_comment_message_en'].apply(clean_and_normalize_text)

# Lưu lại file CSV đã làm sạch
cleaned_file_path = '/content/drive/MyDrive/1_Python_practices/final_project_data/cleaned_file.csv'
df.to_csv(cleaned_file_path, index=False, encoding='utf-8')
```
